## SVM and SMO

## SVM: Large Margin classifier
`支持向量机(Support Vector Machine，SVM)`是一种功能强大的机器学习算法，广泛用于线性和非线性分类，以及回归和离群点检测任务，在解决小样本、非线性及高维模式识别中表现出许多特有的优势。本实验主要关注SVM在分类问题上的应用。下图显示了A、B、C三个样本以及对应的分类面。容易看出A点远离决策边界，因此可以“很有把握的”将其预测为正样本，而C点非常靠近决策边界，尽管仍可以将其预测为正样本，但发生一点小扰动就会使其跑到决策面的另一边从而被识别成负样本。因此可以说我们在预测A为正样本的confidence远高于预测C。

### 假设函数hypothesis
重新定义符号如下：

类别标签 $y=\{0,1\} \Rightarrow y=\{-1,+1\}$

偏置 $\theta_0  \Rightarrow b$

权重 $w = \left[ 
\begin{array}{c}
w_1 \\
w_2 \\
\vdots \\
w_n
\end{array} 
\right] = \left[ 
\begin{array}{c}
\theta_1 \\
\theta_2 \\
\vdots \\
\theta_n
\end{array} 
\right]$


SVM的假设函数如下 $h_{w,b}(x) = w^Tx + b$

分类规则 $y  = \mathrm{sign}(w^T x+ b) = \left\{ \begin{array}{ll} +1,  & w^T x+ b > 0 \\
        -1, & w^T x+ b < 0 \end{array} \right.$

分类面方程 $\theta^T x = 0  \Rightarrow w^T x + b = 0$

正确分类的条件：
$
y_i (w^T x_i + b) >0
$




### Two kind of margins

#### 函数间隔 **functional margin**:
定义第$i$个样本$({x^{(i)}},{y^{(i)}})$的函数间隔为：

$${{\hat \gamma }^{(i)}} = {y^{(i)}}({w^T}{x^{(i)}} + b)$$

对于所有样本 $S = \{ ({x^{(i)}},{y^{(i)}}),i = 1,...,m\} $, 可定义其函数间隔为所有样本函数间隔最小值:

$$\hat \gamma  = \mathop {\min}\limits_{i = 1,...,m} {{\hat \gamma }^{(i)}}$$

- 但这样定义的函数间隔有问题，即如果成比例的改变$w$和$b$（如将它们改成$2w$和$2b$），则函数间隔的值却变成了原来的2倍（虽然此时超平面没有改变），所以只有函数间隔还不够，我们要引入几何间隔。

#### 几何间隔 **geometric margin** 

几何学里，点($x$，$y$)到直线$ax+by+c=0$的距离公式:
$$d(x, y) = \frac{|ax+by+c|}{\sqrt{a^{2}+b^{2}}}$$
所以在二维空间中，几何间隔就是点到直线的距离；在三维空间中，几何间隔是点到空间上一平面的距离；而在SVM中，平面变成了超平面，几何间隔便是样本点$(x^{(i)},y^{(i)})$到超平面的距离：


$${\gamma ^{(i)}} = {y^{(i)}}(\frac{{{w^T}}}{{||w||}} \cdot x^{(i)} + \frac{b}{{||w||}})$$

同样可定义集合$S$的几何间隔:

$$\gamma  = \mathop {\min}\limits_{i = 1,...,m} {\gamma ^{(i)}}$$

显然$||w||=1$ 两者等价 

$${\gamma ^{(i)}} = \frac{{{{\hat \gamma }^{(i)}}}}{{||w||}}$$

$$\gamma  = \frac{{\hat \gamma }}{{||w||}}$$



### Maximizing margin 最大间隔分类器
如下图所示，中间的实线便是寻找到的 `最优超平面（Optimal Hyper Plane）`，其到两条虚线边界的距离相等，虚线边界上的点到最优超平面的距离便是几何间隔，两条虚线间隔边界之间的距离等于2倍的几何间隔，而虚线间隔边界上的点则是支持向量。而对于所有不是支持向量的点，有如下关系：<br>
$$y(w^Tx+b)\geq 1$$
<div align=center><img src="images/3.png" style="zoom:50%" ></div>

函数间隔不适合用来最大化间隔值，因为在超平面固定以后，可以等比例地缩放$w$的长度和$b$的值，这样可以使得的值任意大，亦即函数间隔可以在超平面保持不变的情况下被取得任意大。但几何间隔因为除上了${\Vert w\Vert}$，使得在缩放$w$和$b$的时候几何间隔的值是不会改变的，它只随着超平面的变动而变动，因此，这是更加合适的一个间隔。换言之，这里要找的最大间隔分类超平面中的“间隔”指的是几何间隔。 于是最大间隔分类器（maximum margin classifier）的目标函数在满足一些条件，根据间隔的定义，可以定义为： 
$$y^{(i)}(w^T x^{(i)}+b)=\hat\gamma_i \geq \hat\gamma,i=1,2,...,m$$
距离超平面最近的这几个训练样本点使得等式成立，它们被称为 **`支持向量`**，这些支持向量到到超平面的距离是$\frac{1}{w}$
从而上述目标函数转化成了：
$$\max\frac{1}{\Vert w\Vert},s.t.\ y^{(i)}(w^Tx^{(i)}+b) \geq 1,i=1,2,...,m$$

相当于在相应的约束条件下，最大化这个$\frac{1}{\Vert w\Vert}$值，而$\frac{1}{\Vert w\Vert}$便是要求解的几何间隔。

### 上面通过引入函数间隔和几何间隔的定义来推出优化目标函数，大家也可以参考课件上的直接从几何上进行margin推导：
给定边界上的正样本$x^+$和负样本$x^-$，可容易定义$\mathrm{margin} = \| x^+ - x^- \|$, 注意到 $w^T x^+ + b = +1 $，$w^T x^- + b = -1 $， 而向量$w$是超平面$w^T x + b = 0 $的法向量， 因此$x^+ = x^- + \lambda w $， $w^T (x^+ - x^-) = 2 $, $\lambda = \frac{2}{w^T w}$， 因此$\mathrm{margin} = \| x^+ - x^- \| = \|\lambda w \| = \frac{2}{\|w\|}$

注意到最大化 $\frac{1}{{||w||}}$ 等价于最小化 $||w||^2$， 所以上述目标函数等价于：
$$\mathop {\min }\limits_{w,b} \frac{1}{2}||w|{|^2}$$

$$s.t. \ {y^{(i)}}({w^T}x + b) \ge 1,i = 1,...,m$$


## Lagrange对偶（The Lagrange Duality ）

### The primal problem

为了求解该优化问题，可通过拉格朗日对偶性（Lagrange Duality）变换到对偶变量 (dual variable) 的优化问题，即通过求解与原问题等价的对偶问题（dual problem）得到原始问题的最优解. 考虑如下 ***primal problem***:

$$\mathop {\min }\limits_w f(w)$$

$$s.t.~~~~~~{g_i}(w) \le 0,i = 1,...,k$$

$$~~~~~~~~{h_i}(w) = 0,i = 1,...,l$$

Lagrangian函数为:

$$L(w,\alpha ,\beta ) = f(w) + \sum\limits_{i = 1}^k {{\alpha _i}{g_i}(w)}  + \sum\limits_{i = 1}^l {{\beta _i}{h_i}(w)} $$

$\alpha_i$'s 和 $\beta_i$'s 为Lagrangian函数为乘子. 令

$${\theta _P}(w) = \mathop {\max }\limits_{\alpha ,\beta :{\alpha _i} \ge 0} L(w,\alpha ,\beta )$$

容易验证，当某个约束条件不满足时，显然有$\theta\left(w\right)=\infty$（只要令$\alpha_i=\infty$即可，而当所有约束条件都满足时，则最优值为$\theta\left(w\right)=\frac{1}{2}{\|w\|}^2$， 即

$$
\theta_P(w) = 
\begin{cases}
f(w), & \text{if } w \text{ satisfies the primal constraints} \\
\infty, & \text{otherwise}
\end{cases}
$$

因此如下最小化问题等价于原始 primal problem

$$\mathop {\min }\limits_w f(w) = \mathop {\min }\limits_w {\theta _P}(w) = \mathop {\min }\limits_w \mathop {\max }\limits_{\alpha ,\beta :{\alpha _i} \ge 0} L(w,\alpha ,\beta )$$

可定义primal problem的最优解对应的值为 ${p^ * } = \mathop {\min }\limits_w {\theta _P}(w)$. 

### The dual problem

再来考虑下对偶问题，令

$${\theta _D}(\alpha ,\beta ) = \mathop {\min }\limits_w L(w,\alpha ,\beta )$$

对偶问题 ***dual problem*** 可以表示成

$$\mathop {\max }\limits_{\alpha ,\beta :{\alpha _i} \ge 0} {\theta _D}(\alpha ,\beta ) = \mathop {\max }\limits_{\alpha ,\beta :{\alpha _i} \ge 0} \mathop {\min }\limits_w L(w,\alpha ,\beta )$$

注意到与原primal 问题的区别仅在于最小最大的顺序不同。同样可定义对偶problem的最优解对应的值${d^ * } = \mathop {\max }\limits_{\alpha ,\beta :{\alpha _i} \ge 0} {\theta _D}(\alpha ,\beta )$.

### Relationship between the primal and the dual

两者最优解对应的值满足：

$${d^ * } = \mathop {\max }\limits_{\alpha ,\beta :{\alpha _i} \ge 0} \mathop {\min }\limits_w L(w,\alpha ,\beta ) \le \mathop {\min }\limits_w \mathop {\max }\limits_{\alpha ,\beta :{\alpha _i} \ge 0} L(w,\alpha ,\beta ) = {p^ * }$$

**证明:**  $\forall w,\alpha ,\beta $, 有 $\mathop {\min }\limits_w L(w,\alpha ,\beta ) \le L(w,\alpha ,\beta ) \le \mathop {\max }\limits_{\alpha ,\beta :{\alpha _i} \ge 0} L(w,\alpha ,\beta )$; 因此 ${\theta _D}(\alpha ,\beta ) \le {\theta _P}(w)$; 可得 $\mathop {\max }\limits_{\alpha ,\beta :{\alpha _i} \ge 0} {\theta _D}(\alpha ,\beta ) \le \mathop {\min }\limits_w {\theta _P}(w)$ 

强对偶性：Under certain conditions, we have ${d^ * } = {p^ * }$:

* $f$ are convex
* $h_i$'s are affine
* $g_i$'s are convex set (A subset $C$ is convex if, for all $x$ and $y$ in $C$, the line segment connecting $x$ and $y$ is included in $C$.) 

在这种情况下，我们可以通过求解对偶问题来间接求解原始问题。

Under these conditions, there must exists ${w^ * }$, ${\alpha ^ * }$, ${\beta ^ * }$: ${w^ * }$ is the solution for the primal problem; ${\alpha ^ * }$, ${\beta ^ * }$ are solutions to the dual problem. More over, ${d^ * } = {p^ * } = L({w^ * },{\alpha ^ * },{\beta ^ * })$.

***Karush-Kuhn-Tucker (KKT) 条件:*** given the above conditions, ${w^ * }$, ${\alpha ^ * }$, ${\beta ^ * }$ are solutions for the primal and dual problems is the sufficient and necessary condition that ${w^ * }$, ${\alpha ^ * }$, ${\beta ^ * }$ satisfy the KKT conditions:

$$\frac{\partial }{{\partial {w_i}}}L({w^*},{\alpha ^*},{\beta ^*}) = 0,\;i = 1,...,n$$

$$\frac{\partial }{{\partial {\alpha _i}}}L({w^*},{\alpha ^*},{\beta ^*}) = 0,\;i = 1,...,k$$

$$\frac{\partial }{{\partial {\beta _i}}}L({w^*},{\alpha ^*},{\beta ^*}) = 0,\;i = 1,...,l$$

$$\alpha _i^*{g_i}({w^*}) = 0,\;i = 1,...,k$$

$${g_i}({w^*}) \le 0,\;i = 1,...,k$$

$$\alpha _i^* \ge 0,\;i = 1,...,k$$

$${h_i}({w^*}) = 0,\;i = 1,...,l$$

第四个式子称为 ***KKT dual complementary*** 条件. If $\alpha _i^* > 0$, then ${g_i}({w^*}) = 0$. 这表明支持向量机（SVM）只依赖于少量的支持向量。同时，也为我们提供了SMO中的收敛条件。

## Optimize the margin

### Hard SVM

如果样本是线性可分的, 则对应的primal问题为

$$\mathop {\min }\limits_{w,b} \frac{1}{2}||w|{|^2}$$

$$s.t.\; - {y^{(i)}}({w^T}x + b) + 1 \le 0,\;i = 1,...,m$$

这是个标准的凸二次规划问题 ***convex QP problem***，可通过标准的软件求解，这里我们讨论采用求解其对偶形式来解决此问题。首先构造其拉格朗日函数：

$$L(w,b,\alpha ) = \frac{1}{2}||w|{|^2} - \sum\limits_{i = 1}^m {{\alpha _i}[{y^{(i)}}({w^T}{x^{(i)}} + b) - 1]} $$

对应的对偶问题如下

$$\mathop {\max }\limits_\alpha  {\theta _D}(\alpha ) = \mathop {\max }\limits_\alpha  \mathop {\min }\limits_{w,b} L(w,b,\alpha )$$

首先, $L$ 对$w$ 和 $b$分别求导，令对应的倒数为零，可得到 $\mathop {\min }\limits_{w,b} L(w,b,\alpha )$:

$${\nabla _w}L(w,b,\alpha ) = w - \sum\limits_{i = 1}^m {{\alpha _i}{y^{(i)}}{x^{(i)}}}  = 0$$

$$\frac{\partial }{{\partial b}}L(w,b,\alpha ) =  - \sum\limits_{i = 1}^n {{\alpha _i}{y^{(i)}}}  = 0$$

容易有 $w = \sum\limits_{i = 1}^m {{\alpha _i}} {y^{(i)}}{x^{(i)}} $和$\sum\limits_{i = 1}^n {{\alpha _i}{y^{(i)}}}  = 0$. 将其带入 $L(w,b,\alpha )$:

$$L(w,b,\alpha ) = \sum\limits_{i = 1}^m {{\alpha _i}}  - \frac{1}{2}\sum\limits_{i,j = 1}^m {{y^{(i)}}{y^{(j)}}{\alpha _i}{\alpha _j}\left\langle {{x^{(i)}},{x^{(j)}}} \right\rangle } $$

因此对应的对偶问题可化为:

$$\mathop {\max }\limits_\alpha  \;W(\alpha ) = \sum\limits_{i = 1}^m {{\alpha _i}}  - \frac{1}{2}\sum\limits_{i,j = 1}^m {{y^{(i)}}{y^{(j)}}{\alpha _i}{\alpha _j}\left\langle {{x^{(i)}},{x^{(j)}}} \right\rangle } $$

$$s.t.\quad {\alpha _i} \ge 0,\;i = 1,...,m$$

$$\sum\limits_{i = 1}^n {{\alpha _i}{y^{(i)}}}  = 0$$

这里$\left<\cdot, \cdot \right>$表示内积。这仍然是一个二次规划问题，可采用通用的QP算法求解，但其规模正比于训练样本数$m$


### Soft SVM

在原先的SVM模型中，outlier的存在可能造成很大的影响，因为超平面本身就是只有少数几个support vector组成，如果这些support vector里又存在outlier的话，其影响就很大了。例如下图：

<div align=center><img src="images/4.png" style="zoom:80%" ></div>

如果直接忽略左上角outlier的话，原来的分隔超平面还是挺好的，但是由于这个 outlier 的出现，导致分隔超平面不得不被挤歪了，变成途中黑线所示（这只是一个示意图，并没有严格计算精确坐标），同时 margin 也相应变小了。当然，更严重的情况是，如果这个 outlier 再往右上移动一些距离的话，我们将无法构造出能将数据分开的超平面来。为了处理这种情况，SVM 允许数据点在一定程度上偏离一下超平面。考虑outlier的问题，我们原来的约束条件由原来的
$$y_i(w^Tx_i+b) \geq 1,~~~i=1,...,m$$
变成了
$$y_i(w^Tx_i+b) \geq 1-\xi_i,~~~i=1,...m$$ 

其中$\xi_i \geq 0$称为松弛变量，对应数据点$x_i$允许偏离的量 [$\xi_i = \max( 1 - {y^{(i)}}({w^T}{x^{(i)}} + b), 0)$]。当然，如果我们允许$\xi_i$任意大的话，那任意的超平面都是符合条件的了。所以，我们在原来的目标函数后面加上一项，使得这些$\xi_i$的总和也要最小：
$$\min\frac{1}{2}{\parallel w\parallel}^2+C\sum_{i=1}^{m}\xi_i$$

其中$C$是一个参数，用于控制目标函数中两项("寻找margin最大的超平面和保证数据点偏差量最小"）之间的权重。注意，其中$\xi$是需要优化的变量，而$C$是一个事先确定好的常量。完整地写出来是这个样子：


$$\mathop {\min }\limits_{w,b} \frac{1}{2}||w|{|^2} + C\sum\limits_{i = 1}^m {{\xi _i}} $$

$$s.t.\;{y^{(i)}}({w^T}{x^{(i)}} + b) \ge 1 - {\xi _i},\;i = 1,...,m$$

$${\xi _i} \ge 0,i = 1,...,m$$

拉格朗日函数则为

$$L(w,b,\xi ,\alpha ,\gamma ) = \frac{1}{2}||w|{|^2} + C\sum\limits_{i = 1}^m {{\xi _i}}  - \sum\limits_{i = 1}^m {{\alpha _i}[{y^{(i)}}({w^T}{x^{(i)}} + b) - 1 + {\xi _i}]}  - \sum\limits_{i = 1}^m {{\gamma _i}{\xi _i}} $$

同样的，令其微分为零，有

$${\nabla _w}L(w,b,\xi ,\alpha ,\gamma ) = w - \sum\limits_{i = 1}^m {{\alpha _i}{y^{(i)}}{x^{(i)}}}  = 0$$

$$\frac{\partial }{{\partial b}}L(w,b,\xi ,\alpha ,\gamma ) =  - \sum\limits_{i = 1}^m {{\alpha _i}{y^{(i)}}}  = 0$$

$$\frac{\partial }{{\partial {\xi _i}}}L(w,b,\xi ,\alpha ,\gamma ) = C - {\alpha _i} - {\gamma _i} = 0$$

将其带入拉格朗日函数，可得如下优化问题

$$\mathop {\max }\limits_\alpha  \;W(\alpha ) = \sum\limits_{i = 1}^m {{\alpha _i}}  - \frac{1}{2}\sum\limits_{i,j = 1}^m {{y^{(i)}}{y^{(j)}}{\alpha _i}{\alpha _j}\left\langle {{x^{(i)}},{x^{(j)}}} \right\rangle } $$

$$s.t.\quad 0 \le {\alpha _i} \le C,\;i = 1,...,m$$

$$\sum\limits_{i = 1}^n {{\alpha _i}{y^{(i)}}}  = 0$$

可以看到唯一区别就是现在dual variable $\alpha$多了一个上限$C$。  

### 核函数 

**在以上的问题中，我们都假设数据本身都是线性可分的，事实上，大部分时候数据并不是线性可分的，这个时候满足这样的条件的超平面就根本不存在。在前面，我们已经了解到SVM处理线性可分的情况，那对于非线性的数据SVM如何处理？这时候SVM的处理方法是选择一个核函数$k(\cdot, \cdot)$隐式将数据映射到高纬空间，来解决在原始空间中线性不可分的问题。**


具体来说，在线性不可分的情况下，支持向量机首先在低维空间中完成计算，然后通过核函数将输入空间映射到高维特征空间，最终在高维特征空间中构造出最优分离超平面，从而把平面上本身不好分的非线性数据分开。如图所示，一维数据在二维空间无法划分，从而映射到三维空间里划分：
<div align=center><img src="images/7.png" style="zoom:50%" ></div>

那么核函数$k(\cdot, \cdot)$是如何达到这个效果的呢？对于一个2维空间数据点$v = (x, y)$，要把它映射到3维空间，其中一种映射关系是：$\phi(x, y) = (x^2, \sqrt 2xy, y^2)$。假如有任意两个数据点：$v_1=\boldsymbol{(x_1,y_1)}$, $v_2=\boldsymbol{(x_2,y_2)}$，我们可以直接计算它们对应向量的内积为：

$$
\langle \phi(v_1), \phi(v_2) \rangle = \langle (x_1^2, \sqrt 2x_1 y_1, y_1^2), (x_2^2, \sqrt 2x_2 y_2, y_2^2) \rangle
$$

很明显，这是一个3维运算。假如3维还是无法线性区分，要映射到$m$维去，那这里就是$m$维运算，$m$越大计算量越大。有没有办法能够降低这个计算量呢？我们来展开推导一下：

$$
\langle \phi(v_1), \phi(v_2) \rangle = \langle (x_1^2, \sqrt 2x_1y_1, y_1^2), (x_2^2, \sqrt 2x_2y_2, y_2^2) \rangle = (x_1x_2 + y_1y_2)^2 = (\langle v_1, v_2 \rangle)^2
$$

经过推导可以看到，两个数据点在映射后的向量内积等于映射前向量内积的平方。如果我们引入核函数：$K(v_1, v_2) = \langle \phi(v_1), \phi(v_2) \rangle = (\langle v_1, v_2 \rangle)^2$，那么就可以通过核函数的计算来等价于映射后的向量内积，实现了高维向量运算转换为低维向量计算（本例中是2维向量运算代替了3维向量运算）。

仍然回到我们的SVM中来，假设映射为: $x \rightarrow \phi(x)$，可以直接定义核函数来计算内积:
$
K(x,z) =\langle  \phi(x),\phi(z) \rangle =  \phi(x)^T \phi(z)
$。因此对应的SVM分类规则为:
$$ y  = \mathrm{sign} \left( \sum_{i=1}^m \alpha_i y^{(i)}  \langle \phi(x^{(i)}), \phi(x) \rangle + b \right) 
= \mathrm{sign} \left( \sum_{i=1}^m \alpha_i y^{(i)}  K(x^{(i)}, x) + b \right)$$

对应的优化问题为
$$\mathop {\max }\limits_\alpha  \;W(\alpha ) = \sum\limits_{i = 1}^m {{\alpha _i}}  - \frac{1}{2}\sum\limits_{i,j = 1}^m {{y^{(i)}}{y^{(j)}}{\alpha _i}{\alpha _j} K({x^{(i)}}, {x^{(j)}})  } $$

$$s.t.\quad 0 \le {\alpha _i} \le C,\;i = 1,...,m$$

$$\sum\limits_{i = 1}^n {{\alpha _i}{y^{(i)}}}  = 0$$

## SMO (Sequential minimal optimization)算法 

SMO将该带约束的优化问题分解为一系列可能的最简单子问题，然后通过分析方法解决这些子问题。具体而言SMO采用坐标上升法***coordinate ascent***求解这些乘子。传统的坐标上升法基本思想：每次迭代只沿着一个坐标方向进行优化，固定其他坐标。通过循环优化每个坐标，逐步逼近最优解。
<div align=center><img src="images/coordinateDescent.png" style="zoom:25%" ></div>

对$\alpha_i$而言，由于存在约束$\sum_{i=1}^{m}y_i\alpha_i=0$，因此每次从中任意抽取两个乘子进行优化求解。不失一般性假设每次选出的乘子为$\alpha_1$和$\alpha_2$，固定其它乘子${\alpha_3,\cdots,\alpha_m}$，使得目标函数只是关于$\alpha_1$和$\alpha_2$的函数。这样，不断的从一堆乘子中任意抽取两个求解，不断迭代求解子问题，最终达到求解原问题的目的。

总体上，SMO包含如下2个部分: 
- 仅包含两个拉格朗日乘子$\alpha_1$和$\alpha_2$的QP问题求解方法
- 如何选择两个拉格朗日乘子$\alpha_1$和$\alpha_2$的启发式方法

因为$\sum\limits_{i = 1}^n {{\alpha _i}{y^{(i)}}}  = 0$，固定其它乘子${\alpha_3,\cdots,\alpha_m}$，因此有

$${\alpha _1}{y^{(1)}} + {\alpha _2}{y^{(2)}} =  - \sum\limits_{i = 3}^n {{\alpha _i}{y^{(i)}}}  = \zeta $$

两个因子不好同时求解，所以可先求第二个乘子$\alpha_2$的解（$\alpha_2^{new}$）,得到$\alpha_2$的解之后，再利用$\alpha_2$的解表示$\alpha_1$的解（$\alpha_1^{new}$）。先用${\alpha _2}$ 来表示${\alpha _1}$:

$${\alpha _1} = {y^{(1)}}\zeta  - {\alpha _2}{y^{(1)}}{y^{(2)}}$$

将其带入到目标函数，可得到关于$\alpha_2$的二次函数, 从而转化成关于${\alpha _2}$的优化问题。 为了方便先定义如下notations:

$$f(x) = \sum\limits_{i = 1}^n {{\alpha _i}{y^{(i)}}K({x^{(i)}},x)}  + b$$


$${E_i} = f({x^{(i)}}) - {y^{(i)}},~~K_{ij} = K(x^{(i)}, x^{(j)})$$

$${v_i} = \sum\limits_{j = 3}^n {{\alpha _j}{y^{(j)}}{K_{ij}}}  = f({x^{(i)}}) - \sum\limits_{j = 1}^2 {{\alpha _j}{y^{(j)}}{K_{ij}}}  - b$$

$$\eta  = 2{K_{12}} - {K_{11}} - {K_{22}}$$

对偶问题可化为:

$$W({\alpha _2}) = \sum\limits_{i = 1}^m {{\alpha _i}}  - \frac{1}{2}\sum\limits_{i,j = 1}^m {{y^{(i)}}{y^{(j)}}{\alpha _i}{\alpha _j}{K_{ij}}} $$

$$ = {\alpha _1} + {\alpha _2} - \frac{1}{2}{K_{11}}\alpha _1^2 - \frac{1}{2}{K_{22}}\alpha _2^2 - {y^{(1)}}{y^{(2)}}{K_{12}}{\alpha _1}{\alpha _2} - {y^{(1)}}{\alpha _1}\sum\limits_{i = 3}^m {{\alpha _i}{y^{(i)}}{K_{i1}}}  - {y^{(2)}}{\alpha _2}\sum\limits_{i = 3}^m {{\alpha _i}{y^{(i)}}{K_{i2}}}  + c$$

$$ = {\alpha _1} + {\alpha _2} - \frac{1}{2}{K_{11}}\alpha _1^2 - \frac{1}{2}{K_{22}}\alpha _2^2 - {y^{(1)}}{y^{(2)}}{K_{12}}{\alpha _1}{\alpha _2} - {y^{(1)}}{\alpha _1}{v_1} - {y^{(2)}}{\alpha _2}{v_2}$$

$$ = {y^{(1)}}\zeta  - {\alpha _2}{y^{(1)}}{y^{(2)}} + {\alpha _2} - \frac{1}{2}{K_{11}}{(\zeta  - {\alpha _2}{y^{(2)}})^2} - \frac{1}{2}{K_{22}}\alpha _2^2 - {y^{(2)}}{K_{12}}(\zeta  - {\alpha _2}{y^{(2)}}){\alpha _2} - {v_1}(\zeta  - {\alpha _2}{y^{(2)}}) - {y^{(2)}}{\alpha _2}{v_2}$$

$$ = \frac{1}{2}\eta \alpha _2^2 + ({y^{(2)}}(E_1^{old} - E_2^{old}) - \eta \alpha _2^{old}){\alpha _2} + const.$$

对应的一阶微分和二阶微分为：
$$\frac{{dW({\alpha _2})}}{{d{\alpha _2}}} = \eta {\alpha _2} + ({y^{(2)}}(E_1^{old} - E_2^{old}) - \eta \alpha _2^{old})$$

$$\frac{{{d^2}W({\alpha _2})}}{{d\alpha _2^2}} = \eta $$

注意到 $\eta \le 0$.

*证明: $\eta  = 2{K_{12}} - {K_{11}} - {K_{22}}$ $ = 2\phi {({x^{(1)}})^T}\phi ({x^{(2)}}) - \phi {({x^{(1)}})^T}\phi ({x^{(1)}}) - \phi {({x^{(2)}})^T}\phi ({x^{(2)}})$ $ =  - ||\phi ({x^{(1)}}) - \phi ({x^{(2)}})||_2^2$ $ \le 0$*

* 若 $\eta \lt 0$, 令一阶微分等于$0$:

$$\frac{{dW({\alpha _2})}}{{d{\alpha _2}}} = \eta {\alpha _2} + ({y^{(2)}}(E_1^{old} - E_2^{old}) - \eta \alpha _2^{old}) = 0$$

$$\alpha_2^{new,unc} = \alpha _2^{old} - \frac{{{y^{(2)}}(E_1^{old} - E_2^{old})}}{\eta }$$

* 若 $\eta = 0$, 则该函数为单调函数，在端点($\alpha$ constraints)取得最大值.



接下来讨论下端点， 假设$\alpha_2$的上下边界分别为$H$和$L$，那么有：
$$L \leq \alpha_2^{new} \leq H$$

回忆下 $\alpha$的更新必须满足如下约束条件

$$0 \le {\alpha _2} \le C$$

$$0 \le {\alpha _1} \le C$$

$${\alpha _1}{y^{(1)}} + {\alpha _2}{y^{(2)}} = \zeta $$

当$y^{(1)} \neq y^{(1)} $，注意到$y^{(1)} y^{(1)} = y^{(2)} y^{(2)} = 1,~y^{(1)} y^{(2)} = -1$，根据$\alpha_1^{new}y^{(1)}+\alpha_2^{new}y^{(2)}={\alpha_1}^{old}y^{(1)}+{\alpha_2}^{old}y^{(2)}=\zeta$，可得${\alpha_1}^{old}-{\alpha_2}^{old} = y^{(1)} \zeta =\gamma$。

* 当$\gamma > 0$, $L = 0, ~H=C-\gamma$
* 当$\gamma < 0$, $L = -\gamma,~ H = C$

所以$L=\max(0, -\gamma),~H=\min(C, C -\gamma)$ <br> 如下图所示：<br>
<div align=center><img src="images/5.png" style="zoom:50%" ></div>

同理，当$y^{(1)}=y^{(2)}$时，同样根据$\alpha_1^{new}y^{(1)}+\alpha_2^{new}y^{(2)}={\alpha_1}^{old}y^{(1)}+{\alpha_2}^{old}y^{(2)} = \zeta$，可得：${\alpha_1}^{old}+{\alpha_2}^{old}=\gamma$， 所以$L=\max(0,\gamma - C),H=\min(C,\gamma)$ <br> 如下图所示：<br>
<div align=center><img src="images/6.png" style="zoom:50%" ></div>

即 $\alpha _2$ 应满足如下约束

* if $y^{(1)} \ne y^{(2)}$, $L \le \alpha _2^{new} \le H$，其中 $L = \max (0,\alpha _2^{old} - \alpha _1^{old})$， $H = \min (C,C + \alpha _2^{old} - \alpha _1^{old})$
  
* if $y^{(1)} = y^{(2)}$, $L \le \alpha _2^{new} \le H$，其中 $L = \max (0,\alpha _1^{old} + \alpha _2^{old} - C)$， $H = \min (C,\alpha _1^{old} + \alpha _2^{old})$

因此我们需要对 $\alpha _2$ 的值进行剪裁使其满足上述约束，即

$$
\alpha_2^{new} = 
\begin{cases}
H, & \text{if } \alpha_2^{new,unc} > H \\
\alpha_2^{new,unc}, & \text{if } L \leq \alpha_2^{new,unc} \leq H \\
L, & \text{if } \alpha_2^{new,unc} < L
\end{cases}
$$

得到${\alpha_2}^{new}$后，即可利用其来更新$\alpha_1^{new}$

$$\alpha _1^{new} = \alpha _1^{old} + {y^{(1)}}{y^{(2)}}(\alpha _2^{old} - \alpha _2^{new})$$


### 启发式的方法选取 $\alpha_1$ 和 $\alpha_2$

根据Osuna定理, 如果两个拉格朗日乘子中至少有一个违反了 KKT 条件，那么每一次的优化都会增加目标函数的值。因此，收敛是有保证的。为了更快地收敛，SMO采用一种其方式的方法来选取P $\alpha_1$ 和 $\alpha_2$ :

**Step 1. 选择违反KKT条件的拉格朗日乘子作为 $\alpha _2$，可通过如下两种迭代交替进行的：** 

   * *Iteration  1:* One scan over the entire $\alpha$'s to find violating $\alpha$'s. Then go to Iteration 2.

   * *Iteration  2:* Repeatedly scan over the non-border $\alpha$'s to find violating $\alpha$'s until all non-border $\alpha$'s obey the KKT constraints within $\epsilon$ ($10 ^{-3}$). Then go to iteration 1.

   * Iteration ends after one scan over the entire $\alpha$'s and find all $\alpha$'s obey the KKT conditions within $\epsilon$. 

很明显，这种启发式方法希望把注意力集中在非边界 $\alpha$ 上，因为它们更可能违反KKT条件。在优化过程中，边界 $\alpha$ 会停留在边界上，但非边界 $\alpha$ 则会随着其他 $\alpha$ 的更新而移动。所以我们大部分时间都在扫描和调整非边界的 $\alpha$。在所有非边界$\alpha$都满足后，我们再扫描整个向量，寻找由于调整非边界$\alpha$而变得违反KKT的边界$\alpha$。

简单回顾下KKT条件：

* $L(w,b,\xi ,\alpha ,\gamma )$ 对$\xi$求导，令导数为零，有
$$C - {\alpha _i} - {\gamma _i} = 0,\;i = 1,...,m$$

* KKT 对偶互补条件 

$${\alpha _i}( - {y^{(i)}}({w^T}{x^{(i)}} + b) + 1 - {\xi _i}) = 0,\;i = 1,...,m$$

$${\gamma _i}{\xi _i} = 0,\;i = 1,...,m$$

* 原始约束 以及拉格朗日乘子 ${\alpha _i}$, ${\gamma _i}$ 的约束

$$g({x^{(i)}}) =  - {y^{(i)}}({w^T}{x^{(i)}} + b) + 1 - {\xi _i} \le 0,\;i = 1,...,m$$

$$h({x^{(i)}}) =  - {\xi _i} \le 0,\;i = 1,...,m$$

$${\alpha _i} \ge 0,\;i = 1,...,m$$

$${\gamma _i} \ge 0,\;i = 1,...,m$$

*Notice: KKT 对偶互补条件包括两种情况：约束边界上的目标函数优化和约束内部的优化。对于约束边界上的优化，${\alpha _i}$和${\gamma _i}$非零，因为优化还与${g_i}$和${h_i}$等式或不等式有关。对于约束条件内的优化，${\alpha _i}$和${\gamma _i}$被设为零，因为优化与${g_i}$和${h_i}$无关。(参见课件)*

1. 根据 $C - {\alpha _i} - {\gamma _i} = 0$ 和 ${\alpha _i} \ge 0$，${\gamma _i} \ge 0$, 容易有 $0 \le {\alpha _i} \le C$.

2. 若 ${\alpha _i} = 0$, 则有 ${\gamma _i}=C$, 即 ${\xi _i}=0$, 容易有 ${y^{(i)}}({w^T}{x^{(i)}} + b) \ge 1$

3. 若 $0 < {\alpha _i} < C$, 则 ${\xi _i}=0$, 可得 ${y^{(i)}}({w^T}{x^{(i)}} + b) = 1$

4. 若 ${\alpha _i} = C$, 则 $ - {y^{(i)}}({w^T}{x^{(i)}} + b) + 1 - {\xi _i} = 0$ 以及 ${\xi _i} \ge 0$, 因此有 ${y^{(i)}}({w^T}{x^{(i)}} + b) \le 1$

即最优解需要满足上述条件，因此出现如下情况则违背了 KKT 条件:

* ${\alpha _i} < C$ 且 ${y^{(i)}}({w^T}{x^{(i)}} + b) < 1$

* ${\alpha _i} > 0$ 且 ${y^{(i)}}({w^T}{x^{(i)}} + b) > 1$






**Step 2. $\alpha _1$ 的选取**

$\alpha _1$ 的选取是根据: $\Delta {\alpha _2} = {\alpha_2^{new}} - \alpha_2^{old} =  - \frac{{{y^{(2)}}(E_1^{old} - E_2^{old})}}{\eta }$进行. Platt thought evaluating the kernel function K is time consuming, so SMO approximates the step size by $|E_1^{old} - E_2^{old}|$. 具体而言SMO采用如下步骤选取$\alpha _1$:

* Find a non-border $\alpha _1$ that maximize $|E_1^{old} - E_2^{old}|$. 
* If this does not make enough progress, it starts a sequential scan through the non-border examples, starting at a random position. 
* If this fails too, it starts a sequential scan though all the examples, also starting at a random position.

*注意由于该算法的大部分时间用于调整非边界 $\alpha$ 的值，因此我们常用缓存存储相应的 $E _i$ 值。*

如何更新 $b$ 和 non-border $E$ 缓存？

若 $0 < {\alpha _2} < C$, 根据 KKT 条件, 有 $\sum\limits_{i = 1}^m {{\alpha _i}{y^{(i)}}{K_{i2}}}  + b = {y^{(i)}}$. 因此

$${b^{new} _2} = {y^{(i)}} - \sum\limits_{i = 1}^m {{\alpha _i}{y^{(i)}}{K_{i2}}} = {y^{(i)}} - \sum\limits_{i = 3}^m {{\alpha _i}{y^{(i)}}{K_{i2}}}  - {\alpha _1}{y^{(1)}}{K_{12}} - {\alpha _2}{y^{(2)}}{K_{22}}$$
$$ =  - {E_2} - {y^{(2)}}{K_{22}}(\alpha _2^{new} - \alpha _2^{old}) - {y^{(1)}}{K_{12}}(\alpha _1^{new} - \alpha _1^{old}) + {b^{old}}$$

若 $0 < {\alpha _1} < C$，同理对于 $b^{new} _1$:

$$b_1^{new} =  - {E_1} - {y^{(1)}}{K_{11}}(\alpha _1^{new} - \alpha _1^{old}) - {y^{(2)}}{K_{12}}(\alpha _2^{new} - \alpha _2^{old}) + {b^{old}}$$

因此$b^{new}$ 采用如下形式更新:

* If $0 < {\alpha _2} < C$, $b^{new} = b^{new} _2$;
* Else if $0 < {\alpha _1} < C$, $b^{new} = b^{new} _1$;
* Else $b^{new} = (b^{new} _1 + b^{new} _2) / 2$;

对于每个非边界上的 $\alpha _i$', 对应的 $E _i$ 更新如下:

$$E_i^{new} = \sum\limits_{j = 1}^m {{\alpha _j}{y^{(j)}}{K_{ij}}}  + {b^{new}} - {y^{(i)}}$$

**经过前面几节的铺垫，我们了解SVM的原理和求解方法，现在我们就来用代码实现SVM。**

```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import make_blobs, make_circles, make_moons
from sklearn.preprocessing import StandardScaler
```

```python
class SMOModel:
    """Container object for the model used for sequential minimal optimization."""
    
    def __init__(self, X, y, C, kernel, alphas, b, errors):
        self.X = X               # training data vector
        self.y = y               # class label vector
        self.C = C               # regularization parameter
        self.kernel = kernel     # kernel function
        self.alphas = alphas     # lagrange multiplier vector
        self.b = b               # scalar bias term
        self.errors = errors     # error cache
        self._obj = []           # record of objective function value
        self.m = len(self.X)     # store size of training set
```

分别实现线性核和高斯核
(提示：可采用np.linalg.norm)

```python
def linear_kernel(x, y, b=1):
    """Returns the linear combination of arrays `x` and `y` with
    the optional bias term `b` (set to 1 by default)."""
    ### START THE CODE ###
    if y.ndim == 1:  
       result = np.dot(x, y) + b
    elif y.ndim == 2:  
       result = np.dot(y, x) + b          ### Note the @ operator for matrix multiplication 
    ### END THE CODE  ###    
    return result 


def gaussian_kernel(x, y, sigma=1):
    """Returns the gaussian similarity of arrays `x` and `y` with
    kernel width parameter `sigma` (set to 1 by default)."""
    
    if np.ndim(x) == 1 and np.ndim(y) == 1:                                             # for vector-vector
    ### START THE CODE ###
         result = np.exp(-np.linalg.norm(x - y)**2 / (2 * sigma**2))
    ### END THE CODE  ###           
    elif (np.ndim(x) > 1 and np.ndim(y) == 1) or (np.ndim(x) == 1 and np.ndim(y) > 1):  # for matrix-vector
    ### START THE CODE ###
        result = np.exp(-np.linalg.norm(x - y, axis=1)**2 / (2 * sigma**2))
    ### END THE CODE  ###           
    elif np.ndim(x) > 1 and np.ndim(y) > 1:                                             # for matrix-matrix
    ### START THE CODE ###
        result = np.exp(-np.linalg.norm(x[:, np.newaxis] - y, axis=2)**2 / (2 * sigma**2))
    ### END THE CODE  ###           
    return result
```

实现如下目标函数
$$W({\alpha _2}) = \sum\limits_{i = 1}^m {{\alpha _i}}  - \frac{1}{2}\sum\limits_{i,j = 1}^m {{y^{(i)}}{y^{(j)}}{\alpha _i}{\alpha _j}{K_{ij}}} $$

```python
# Objective function to optimize

def objective_function(alphas, target, kernel, X_train):
    """Returns the SVM objective function based in the input model defined by:
    `alphas`: vector of Lagrange multipliers
    `target`: vector of class labels (-1 or 1) for training data
    `kernel`: kernel function
    `X_train`: training data for model."""
    
   
    ### START THE CODE ###
    n = len(X_train)
    result = 0
    for i in range(n):
        for j in range(n):
            result += target[i] * target[j] * alphas[i]*alphas[j]*kernel(X_train[i], X_train[j])   
            
    result = np.sum(alphas)-1/2*result
    '''n = len(X_train)
    temp = np.array([[target[i] * target[j] * alphas[i] * alphas[j] * kernel(X_train[i], X_train[j]) for j in range(n)] for i in range(n)])
    result = np.sum(alphas) - 1/2 * np.sum(temp) '''
    ### END THE CODE  ###
    return result
```

实现决策函数
$$f(x) = \sum\limits_{i = 1}^n {{\alpha _i}{y^{(i)}}K({x^{(i)}},x)}  - b$$

```python
# Decision function

def decision_function(alphas, target, kernel, X_train, x_test, b):
    """Applies the SVM decision function to the input feature vectors in `x_test`."""
    
    ### START THE CODE ###
    if x_test.ndim == 1:  
       result = np.sum([alphas[i] * target[i] * kernel(X_train[i], x_test) for i in range(len(X_train))]) + b
    else:  
       result = np.array([np.sum([alphas[i] * target[i] * kernel(X_train[i], x) for i in range(len(X_train))]) + b for x in x_test])
    ### END THE CODE  ###
    return result
```

`take_step()` 计算两个新的 $\alpha$ 值、新的阈值 $b$，并更新误差缓存$E_k$。

```python
def take_step(i1, i2, model):
    
    # Skip if chosen alphas are the same
    if i1 == i2:
        return 0, model
    
    alph1 = model.alphas[i1]
    alph2 = model.alphas[i2]
    y1 = model.y[i1]
    y2 = model.y[i2]
    E1 = model.errors[i1]
    E2 = model.errors[i2]
    s = y1 * y2
    
    # Compute L & H, the bounds on new possible alpha values
    if (y1 != y2):
        L = max(0, alph2 - alph1)
        H = min(model.C, model.C + alph2 - alph1)
    elif (y1 == y2):
        L = max(0, alph1 + alph2 - model.C)
        H = min(model.C, alph1 + alph2)
    if (L == H):
        return 0, model

    # Compute kernel & 2nd derivative eta
    k11 = model.kernel(model.X[i1], model.X[i1])
    k12 = model.kernel(model.X[i1], model.X[i2])
    k22 = model.kernel(model.X[i2], model.X[i2])
    eta = 2 * k12 - k11 - k22
    
    # Compute new alpha 2 (a2) if eta is negative
    if (eta < 0):
        a2 = alph2 - y2 * (E1 - E2) / eta
        # Clip a2 based on bounds L & H
        if L < a2 < H:
            a2 = a2
        elif (a2 <= L):
            a2 = L
        elif (a2 >= H):
            a2 = H
            
    # If eta is non-negative, move new a2 to bound with greater objective function value
    else:
        alphas_adj = model.alphas.copy()
        alphas_adj[i2] = L
        # objective function output with a2 = L
        Lobj = objective_function(alphas_adj, model.y, model.kernel, model.X) 
        alphas_adj[i2] = H
        # objective function output with a2 = H
        Hobj = objective_function(alphas_adj, model.y, model.kernel, model.X)
        if Lobj > (Hobj + eps):
            a2 = L
        elif Lobj < (Hobj - eps):
            a2 = H
        else:
            a2 = alph2
            
    # Push a2 to 0 or C if very close
    if a2 < 1e-8:
        a2 = 0.0
    elif a2 > (model.C - 1e-8):
        a2 = model.C
    
    # If examples can't be optimized within epsilon (eps), skip this pair
    if (np.abs(a2 - alph2) < eps * (a2 + alph2 + eps)):
        return 0, model
    
    # Calculate new alpha 1 (a1)
    a1 = alph1 + s * (alph2 - a2)
    
    # Update threshold b to reflect newly calculated alphas
    # Calculate both possible thresholds
    b1 = E1 + y1 * (a1 - alph1) * k11 + y2 * (a2 - alph2) * k12 + model.b
    b2 = E2 + y1 * (a1 - alph1) * k12 + y2 * (a2 - alph2) * k22 + model.b
    
    # Set new threshold based on if a1 or a2 is bound by L and/or H
    if 0 < a1 and a1 < C:
        b_new = b1
    elif 0 < a2 and a2 < C:
        b_new = b2
    # Average thresholds if both are bound
    else:
        b_new = (b1 + b2) * 0.5

    # Update model object with new alphas & threshold
    model.alphas[i1] = a1
    model.alphas[i2] = a2
    
    # Update error cache
    # Error cache for optimized alphas is set to 0 if they're unbound
    for index, alph in zip([i1, i2], [a1, a2]):
        if 0.0 < alph < model.C:
            model.errors[index] = 0.0
    
    # Set non-optimized errors based on above equations
    non_opt = [n for n in range(model.m) if (n != i1 and n != i2)]
    model.errors[non_opt] = model.errors[non_opt] + \
                            y1*(a1 - alph1)*model.kernel(model.X[i1], model.X[non_opt]) + \
                            y2*(a2 - alph2)*model.kernel(model.X[i2], model.X[non_opt]) + model.b - b_new
    
    # Update model threshold
    model.b = b_new
    
    return 1, model
```

`examine_example()`启发式选取第二个$\alpha$，并将两个需要优化的$\alpha$对应的索引传递给`take_step()`函数进行优化


```python
def examine_example(i2, model):
    
    y2 = model.y[i2]
    alph2 = model.alphas[i2]
    E2 = model.errors[i2]
    r2 = E2 * y2

    # Proceed if error is within specified tolerance (tol)
    if ((r2 < -tol and alph2 < model.C) or (r2 > tol and alph2 > 0)):
        
        if len(model.alphas[(model.alphas != 0) & (model.alphas != model.C)]) > 1:
            # Use 2nd choice heuristic is choose max difference in error
            if model.errors[i2] > 0:
                i1 = np.argmin(model.errors)
            elif model.errors[i2] <= 0:
                i1 = np.argmax(model.errors)
            step_result, model = take_step(i1, i2, model)
            if step_result:
                return 1, model
            
        # Loop through non-zero and non-C alphas, starting at a random point
        for i1 in np.roll(np.where((model.alphas != 0) & (model.alphas != model.C))[0],
                          np.random.choice(np.arange(model.m))):
            step_result, model = take_step(i1, i2, model)
            if step_result:
                return 1, model
        
        # loop through all alphas, starting at a random point
        for i1 in np.roll(np.arange(model.m), np.random.choice(np.arange(model.m))):
            step_result, model = take_step(i1, i2, model)
            if step_result:
                return 1, model
    
    return 0, model
```

 函数`train()`启发式选择第一个要优化的 $\alpha$，并将此值传递给 `examine_example()`。该函数使用 while 循环，以几种不同的方式遍历 $alpha$ 值，直到无法再进行优化为止，然后返回优化后的$alpha$向量（嵌入 SMOModel 对象中）。


```python
def train(model):
    
    numChanged = 0
    examineAll = 1

    while(numChanged > 0) or (examineAll):
        numChanged = 0
        if examineAll:
            # loop over all training examples
            for i in range(model.alphas.shape[0]):
                examine_result, model = examine_example(i, model)
                numChanged += examine_result
                if examine_result:
                    obj_result = objective_function(model.alphas, model.y, model.kernel, model.X)
                    model._obj.append(obj_result)
        else:
            # loop over examples where alphas are not already at their limits
            for i in np.where((model.alphas != 0) & (model.alphas != model.C))[0]:
                examine_result, model = examine_example(i, model)
                numChanged += examine_result
                if examine_result:
                    obj_result = objective_function(model.alphas, model.y, model.kernel, model.X)
                    model._obj.append(obj_result)
        if examineAll == 1:
            examineAll = 0
        elif numChanged == 0:
            examineAll = 1
```

```python
def plot_decision_boundary(model, ax, resolution=100, colors=('b', 'k', 'r'), levels=(-1, 0, 1)):
        """Plots the model's decision boundary on the input axes object.
        Range of decision boundary grid is determined by the training data.
        Returns decision boundary grid and axes object (`grid`, `ax`)."""
        
        # Generate coordinate grid of shape [resolution x resolution]
        # and evaluate the model over the entire space
        xrange = np.linspace(model.X[:,0].min(), model.X[:,0].max(), resolution)
        yrange = np.linspace(model.X[:,1].min(), model.X[:,1].max(), resolution)
        grid = [[decision_function(model.alphas, model.y,
                                   model.kernel, model.X,
                                   np.array([xr, yr]), model.b) for xr in xrange] for yr in yrange]
        grid = np.array(grid).reshape(len(xrange), len(yrange))
        
        # Plot decision contours using grid and
        # make a scatter plot of training data
        ax.contour(xrange, yrange, grid, levels=levels, linewidths=(1, 1, 1),
                   linestyles=('--', '-', '--'), colors=colors)
        ax.scatter(model.X[:,0], model.X[:,1],
                   c=model.y, cmap=plt.cm.viridis, lw=0, alpha=0.25)
        
        # Plot support vectors (non-zero alphas)
        # as circled points (linewidth > 0)
        mask = np.round(model.alphas, decimals=2) != 0.0
        ax.scatter(model.X[mask,0], model.X[mask,1],
                   c=model.y[mask], cmap=plt.cm.viridis, lw=1, edgecolors='k')
        
        return grid, ax
```

# 支持向量机回归 Support Vector Regression (SVR)
可将SVM中的hinge 损失换成其他通用的损失进行拓展，得到更一般的形式
$$
\hbox{min}_{w,b}  
 C \sum_{i=1}^m \ell \left(y^{(i)}, h_{w,b}(x^{(i)}) \right) + \frac{1}{2} ||w||^2 
$$

选择合适的损失函数可以将SVM的思想来解决回归问题。传统的回归模型通常直接基于模型输出$h_{w,b}(x)$与真实输出$y$之间的差来计算损失，只有当两者完全相同时，损失才为0。而SVR加入了一个$\epsilon$参数，意为我们可以**容忍最多有$\epsilon$偏差**，如下图
<div align=center><img src="images/9.png" style="zoom:60%" ></div>

也就是说，在虚线之间的部分不进行损失计算，他们的损失为0，而只计算虚线以外的点的损失，因此他的对应损失函数为$\epsilon-$不敏感损失函数$\ell_\epsilon(z)$
$$
 \ell_\epsilon(z) = \left\{ \begin{array}{l c}
  0, & \text{if}~ |z| \le \epsilon; \\
  |z| -\epsilon, & \text{otherwise}
  \end{array} \right.
$$
<div align=center><img src="images/10.png" style="zoom:60%" ></div>

引入松弛因子$ξ_i$、$ξ_i^*$，SVR的优化问题变为：
$$
\begin{align*}
\min_{w,b,\xi_i, \xi_i^*} \quad & C \sum_{i=1}^{m}(\xi_i+\xi_i^*) + \frac{1}{2} \|w\|^2 \\
\text{s.t.} \quad & h_{w,b}(x^{(i)}) - y^{(i)} \leq \epsilon + \xi_i, \\
& y^{(i)} - h_{w,b}(x^{(i)}) \leq \epsilon + \xi_i^*, \\
& \xi_i \geq 0,~\xi_i^* \geq 0,~i = 1, 2, \cdots, m
\end{align*}
$$
其中$h_{w,b}(x)=w^Tx+b$。

同样的，引入拉格朗日乘子,得到对应拉格朗日函数：
$$L(w,b,α,α^*,ξ,ξ^*,r_i,r_i^*)=\frac{1}{2}{\parallel w\parallel}^2+C\sum_{i=0}^{m}(ξ_i+ξ_i^*)-\sum_{i=1}^{m}r_iξ_i-\sum_{i=1}^{m}r_i^*ξ_i^*+\sum_{i=1}^{m}α_i(h_{w,b}(x^i)-y^i-ε-ξ_i)+$$
$$\sum_{i=1}^{m}α_i^*(y^i-h_{w,b}(x^i)-ε-ξ_i^*)$$
令函数中对应偏导为0，得到：
$$
w = \sum_{i=1}^m (\alpha_i^* - \alpha_i) x^{(i)}, ~ 0=\sum_{i=1}^m (\alpha_i^* - \alpha_i), ~
C = \alpha_i + r_i, ~ C = \alpha_i^* + r_i^*
$$

代入到拉格朗日函数中得到之关于$α_i$,$α_i^*$的函数，最大化该函数即得到SVR的对偶问题：

$$
\begin{align*}
 \max_{\alpha, \alpha^*} & \sum_{i=1}^m y^{(i)}(\alpha_i^* - \alpha_i) - \epsilon(\alpha_i^* + \alpha_i) - \frac{1}{2} \sum_{i,j=1}^m (\alpha_i^* - \alpha_i)(\alpha_j^* - \alpha_j)  \langle x^{(i)},  x^{(j)} \rangle \\
 \hbox{s.t.}\; & \;\; 0 \leq \alpha_i,\alpha_i^* \leq C, \;\; i=1,\ldots,m \\
               &  \sum_{i=1}^m (\alpha_i^* - \alpha_i) = 0.
\end{align*}
$$

可以看出其仍为QP问题，KKT条件为：
$$
\left\{
\begin{align*}
 & \alpha_i (h_{w,b}(x^{(i)}) - y^{(i)} - \epsilon - \xi_i) = 0, \\
  &\alpha_i^* (y^{(i)} - h_{w,b}(x^{(i)}) - \epsilon - \xi_i^*) = 0, \\
 & (C-\alpha_i)\xi_i = 0, \\
 & (C-\alpha_i^*)\xi_i^* = 0.
\end{align*}
\right.
$$


因此可以进行求解，同时对于非线性回归也可引入核函数实现。


为了简便，我们这里直接调用scikit-learn中现成的SVR实现来进行实验

```python
#初始化一系列随机样本
np.random.seed(42)
m = 50
X = 2 * np.random.rand(m, 1)
y = (4 + 3 * X + np.random.randn(m, 1)).ravel()  # 将多维数组降为一维

#调用sklearn库，实现LinearSVR
from sklearn.svm import LinearSVR

### START THE CODE ###
svr = LinearSVR(epsilon = 0.3)          #根据 LinearSVR函数 设定epsilon = 0.3,方便观察效果，支持向量均位于容忍区域外侧
svr.fit(X,y)                            #根据 LinearSVR函数 相关文档得到拟合模型
### END THE CODE  ###


#找到支持向量
def find_support(svr,X,y):
    y_pred=svr.predict(X)#计算预测值
    margin=(np.abs(y-y_pred) >= svr.epsilon)#判断是否为支持向量
    return np.argwhere(margin)

svr.support_ = find_support(svr,X,y)#获取对应的支持向量对应下标，注意将变量名改为你自己的命名
    
# 绘制SVR结果
def plot_svr(svr,X,y,axes):
    xls=np.linspace(axes[0],axes[1],100).reshape(100,1)
    y_pred=svr.predict(xls)
    plt.plot(xls, y_pred, "-")
    plt.plot(xls, y_pred + svr.epsilon, "--")  
    plt.plot(xls, y_pred - svr.epsilon, "--")
    plt.plot(X, y, "bo")
    plt.scatter(X[svr.support_], y[svr.support_], s=180, facecolors='#FFAAAA')
    plt.axis(axes)
plt.figure(figsize=(9, 4))  # width, height in inches
plot_svr(svr, X, y, [0, 2, 3, 11]) #注意第一个参数改为自己的命名
```

通过观察图像可以看出，SVR中的支持向量均位于容忍区域外侧

## 非线性SVR

非线性回归中，参数$C$越大，对错样本的惩罚程度越大，正则化项的作用会越小，模型越趋向于过拟合。
反之，$C$越小，正则化效果越强，模型会更简单。
在实验中同学们可以通过调整$C$参数观察回归结果的不同。

```python
#初始化随机样本
np.random.seed(42)
m = 100
X = 2 * np.random.rand(m, 1) - 1
y = (0.2 + 0.1 * X + 0.5 * X**2 + np.random.randn(m, 1)/10).ravel()

#实现SVR
from sklearn.svm import SVR

### START THE CODE ###
svr_poly = SVR(kernel='poly', gamma='auto', C=1e3, degree=4)#设定C = 1e3
svr_poly.fit(X,y) 
### END THE CODE  ###

plt.figure(figsize=(9, 4))
plot_svr(svr_poly, X, y, [-1, 1, 0, 1])
```
